Задание №4. Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

```python
directories = {
    '1': ['2207 876234', '11-2', '5455 028765'],
    '2': ['10006', '5400 028765', '5455 002299'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (Внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):

p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит. 
s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится. Правильно обработайте ситуацию, когда пользователь будет вводить несуществующий документ. 
l – list – команда, которая выведет список всех документов в формате passport "2207 876234" "Василий Гупкин". 
as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. 
d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ. 
m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку. 
a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.

```python
def search_doc(num=''):
    """
    Функция осуществляет обход элементов documents - каталог документов.
    Параметр num - номер документа. 
    Если num заполнено, то функция осуществляет поиск по каталогу документов с выводом имени владельца документа.
    Если num не заполнено, то функция выводит список всех документов каталога.
    """
    
    fl = 0
    str_mes = ''
    inx = ''
    answer = dict()    
   
    for el in documents:
        if num == '': 
            fl = 1
            str = ' '
            for value in el.values():
                str += value + ' '
            str_mes += str.strip() + '\n'
        else:
            if el['number'] == num:
                fl = 1
                str_mes = f'Документ с номером {num} принадлежит: {el["name"]}'
                inx = documents.index(el)
    
    if fl == 0:
        str_mes = f'Документ с номером {num} в каталоге не найден!'
    
    answer['fl'] = fl
    answer['str_mes'] = str_mes
    answer['index'] = inx
    
    return answer


def add_doc(type_doc, num, name_user):
    answer = dict()
    
    documents.append({'type': type_doc, 'number': num, 'name': name_user})
    fl = 1
    str_mes = f'Документ добавлен в каталог: {type_doc} {num} {name_user}'
    
    answer['fl'] = fl
    answer['str_mes'] = str_mes
    
    return answer    


def del_doc(num):
    """
    Функция обработки команды: d – delete–команда, которая спросит номер док-та и удалит его из каталога и из перечня полок
    """
    answer = search_doc(num)
    if answer.get('fl') == 0:
        print_mes(answer.get('str_mes'))
    else:
        eldel = answer.get('index')
        if eldel != '':
            del(documents[eldel])
            print_mes(f'Документ с номером {num} удален из каталога!')
            answer = seach_on_dir(num)
            elkey = answer.get('key')
            ellen = answer.get('len')
            answer = del_dir(num, elkey, ellen)
    
    
def seach_on_dir(num=''):
    """
    Функция осуществляет поиск номера полки, на которой находится документ (поиск в directories - перечень полок).
    Параметр num - номер документа.
    """
    
    fl = 0
    str_mes = ''
    key_dir = ''
    answer = dict()    
    
    if num != '':
        for key,value in directories.items():
            if value != None:
                if num in value:
                    key_dir = key
                    fl = 1
                    str_mes = f'Документ с номером {num} находится на полке {key_dir}'
    if fl == 0:
        str_mes = f'Документ с номером {num} на полках не найден!'
    
    answer['fl'] = fl
    answer['str_mes'] = str_mes
    answer['key'] = key_dir
    
    return answer       

    
def actions_dir(num='', shelf=''):
    """
    Функция добавляет указанный документ на указанную полку (в directories - перечень полок).
    Параметр num - номер документа.
    Параметр shelf - номер полки.
    """
    if num != '' and shelf != '':
        list_value = directories[shelf]
        if list_value != None:
            list_value.append(num)
            directories[shelf] = list_value
        else:
            directories[shelf] = [num]
    print_mes(f'Документ с номером {num} добавлен на полку {shelf}') 

    
def add_dir(shelf):
    """
    Функция добавляет новую полку (новый элемент в directories - перечень полок).
    Параметр shelf - номер полки.    
    """
    
    fl = 0
    str_mes = ''
    answer = dict()    
    
    if shelf != '':
        if shelf not in directories:
            fl = 1
            directories.setdefault(shelf, list())
            str_mes = f'Полка с номером {shelf} добавлена!'
        else:
            str_mes = f'Полка с номером {shelf} уже существует!'
       
    answer['fl'] = fl
    answer['str_mes'] = str_mes
    
    return answer   


def del_dir(num, elkey, ellen):
    """
    Функция удаляет документ с полки и удаяет саму полку, если документ был единственным
    (в directories - перечень полок).    
    """    
    str_mes = ''
    if ellen == 1:
        del directories[elkey]
        str_mes = f'Документ с номером {num} и полка {elkey} удалены!'
    else:
        list_value = directories[elkey]
        list_value.remove(num)
        directories[elkey] = list_value
        str_mes = f'Документ с номером {num} удален с полки {elkey}!'
    print_mes(str_mes)
      
                     
def print_mes(str_mes):
    """
    Функция выводит на экран переданный текст-сообщение    
    """
    print(str_mes)  

    
def put_doc(num, shelf):
    """
    Функция обработки команды: a – add – команда, которая добавит новый документ в каталог и в перечень полок,
    спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.
    """
    answer = add_dir(shelf)
    print_mes(answer.get('str_mes'))
    answer = search_doc(num)
    if answer.get('fl') == 1:
        print_mes(answer.get('str_mes'))
    else:
        type_doc = input('Укажите тип документа: ')
        name_user = input('Укажите имя владельца документа: ')
        answer = add_doc(type_doc, num, name_user)
        print_mes(answer.get('str_mes'))
        answer = actions_dir(num, shelf)    

        
def move_doc(num, shelf):
    """
    Функция обработки команды: m – move–команда, которая спросит номер док-та и целевую полку,
    и переместит его с текущей полки на целевую.
    """        
    answer = search_doc(num)
    if answer.get('fl') == 0:
        print_mes(answer.get('str_mes'))
    else:
        answer = add_dir(shelf)
        if answer.get('fl') == 0:
            #проверить если такой номер на полке
            answer = seach_on_dir(num)
            if answer.get('fl') == 1:
                if answer.get('key') == shelf:
                    print_mes(f'Документ с номером {num} уже находиться на полке {shelf}')
                else:
                    elkey = answer.get('key')
                    ellen = answer.get('len')
                    answer = del_dir(num, elkey, ellen)
                    answer = actions_dir(num, shelf)                
        else:
             #перемещение на вновь созданную полку
            print_mes(answer.get('str_mes'))
            answer = actions_dir(num, shelf)
               

def fun_chief():
    #Основная функция
    
    print('''
        Список команд:
        p - выводит имя владельца документа
        l - выводит список всех документов
        s - выводит номер полки, на которой лежит документ        
        as - добавляет новую полку
        d - удаляет документ из каталога и перечня полок
        a - добавляет новый документ в каталог и перечень полок        
        m - перемещает документ на другую полку
        q - выход
        ''')
  
    while True:    
        command = str(input('Введи команду: ')).lower().strip()

        if command == 'p':
            num = input('Укажите номер документа: ')
            answer = search_doc(num)
            print_mes(answer.get('str_mes'))
        elif command == 'l':
            answer = search_doc()
            print_mes(answer.get('str_mes'))
        elif command == 's':
            num = input('Укажите номер документа: ')
            answer = seach_on_dir(num)
            print_mes(answer.get('str_mes'))    
        elif command == 'as':
            shelf = input('Укажите номер полки для добавления в перечень: ')
            answer = add_dir(shelf)
            print_mes(answer.get('str_mes'))
        elif command == 'a':
            shelf = input('Укажите номер полки, на который нужно добавить новый документ: ')
            num = input('Укажите номер документа: ')
            put_doc(num, shelf)
        elif command == 'd':
            num = input('Укажите номер документа: ')
            if num != '':
                del_doc(num)
            else:
                print_mes('Задайте корректный номер документа!')
        elif command == 'm':
            num = input('Укажите номер документа: ')
            if num != '':
                shelf = input('Укажите номер полки, на которую нужно переместить документ: ')
                if shelf != '': 
                    move(num, shelf)
                else:
                    print_mes('Задайте корректный номер полки!')
            else:
                print_mes('Задайте корректный номер документа!')
        elif command == 'q':
            str_mes = 'Выход из программы!'
            print_mes(str_mes) 
            break
        else:
            str_mes = f'Команда {command} не распознана!'
            print_mes(str_mes) 

    
fun_chief()
```
